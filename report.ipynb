```python
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import TextVectorization

# Отключаем предупреждения
import warnings
warnings.filterwarnings('ignore')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# =============================================================================
# КОНФИГУРАЦИЯ
# =============================================================================

# Путь к датасету на Kaggle
DATASET_PATH = '/kaggle/input/corpus-of-russian-news-articles-from-lenta/lenta-ru-news.csv'

# Параметры данных
SAMPLE_SIZE = 100000 
MAX_TOPICS = 5
TEST_SIZE = 0.2
VAL_SIZE = 0.1

# Параметры нейронной сети
MAX_FEATURES = 10000
SEQUENCE_LENGTH = 200
EMBEDDING_DIM = 128

# Параметры обучения
EPOCHS = 10
BATCH_SIZE = 64

print(f"\nКонфигурация:")
print(f"  - Размер выборки: {SAMPLE_SIZE}")
print(f"  - Количество классов: {MAX_TOPICS}")
print(f"  - Размер словаря: {MAX_FEATURES}")
print(f"  - Длина последовательности: {SEQUENCE_LENGTH}")
print(f"  - Количество эпох: {EPOCHS}")
print(f"  - Размер батча: {BATCH_SIZE}")

# =============================================================================
# 1. ЗАГРУЗКА И ПРЕДОБРАБОТКА ДАННЫХ
# =============================================================================

print("\n" + "="*70)
print("1. ЗАГРУЗКА И ПРЕДОБРАБОТКА ДАННЫХ")
print("="*70)

# Загрузка данных
print(f"\nЗагрузка данных из {DATASET_PATH}...")
df = pd.read_csv(DATASET_PATH, nrows=SAMPLE_SIZE)
print(f"Загружено {len(df)} записей")

# Удаление пропусков
df = df.dropna(subset=['text', 'topic'])
print(f"После удаления пропусков: {len(df)} записей")

# Распределение по темам
print("\nРаспределение по темам:")
print(df['topic'].value_counts())

# Выбор топ-N тем
top_topics = df['topic'].value_counts().head(MAX_TOPICS).index.tolist()
df_filtered = df[df['topic'].isin(top_topics)].copy()

print(f"\nИспользуем {MAX_TOPICS} самых популярных тем:")
for i, topic in enumerate(top_topics, 1):
    count = len(df_filtered[df_filtered['topic'] == topic])
    print(f"  {i}. {topic}: {count} записей")

# Очистка текста
print("\nОчистка текстов...")

def clean_text(text):
    """Очистка текста"""
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^а-яё\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df_filtered['cleaned_text'] = df_filtered['text'].apply(clean_text)
df_filtered = df_filtered[df_filtered['cleaned_text'].str.len() > 0]

# Кодирование меток
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(df_filtered['topic'])
texts = df_filtered['cleaned_text'].values
class_names = label_encoder.classes_

print(f"Всего обработано {len(texts)} текстов")
print(f"Классы: {list(class_names)}")

# Разделение данных
print("\nРазделение на выборки...")
X_temp, X_test, y_temp, y_test = train_test_split(
    texts, labels, test_size=TEST_SIZE, random_state=42, stratify=labels
)

val_size_adjusted = VAL_SIZE / (1 - TEST_SIZE)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=val_size_adjusted, random_state=42, stratify=y_temp
)

print(f"  Обучающая: {len(X_train)}")
print(f"  Валидационная: {len(X_val)}")
print(f"  Тестовая: {len(X_test)}")

# =============================================================================
# 2. СОЗДАНИЕ НЕЙРОННОЙ СЕТИ
# =============================================================================

print("\n" + "="*70)
print("2. СОЗДАНИЕ НЕЙРОННОЙ СЕТИ")
print("="*70)

# Векторизатор
print("\nСоздание векторизатора...")
vectorizer = TextVectorization(
    max_tokens=MAX_FEATURES,
    output_mode='int',
    output_sequence_length=SEQUENCE_LENGTH
)
vectorizer.adapt(X_train)
print(f"Размер словаря: {len(vectorizer.get_vocabulary())}")

# Создание модели
print("\nСоздание модели LSTM...")

inputs = keras.Input(shape=(1,), dtype=tf.string)
x = vectorizer(inputs)
x = layers.Embedding(MAX_FEATURES, EMBEDDING_DIM)(x)
x = layers.Dropout(0.3)(x)
x = layers.LSTM(128, return_sequences=True)(x)
x = layers.Dropout(0.3)(x)
x = layers.LSTM(64)(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(MAX_TOPICS, activation='softmax')(x)

model = keras.Model(inputs=inputs, outputs=outputs)

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nАрхитектура модели:")
model.summary()

# =============================================================================
# 3. ОБУЧЕНИЕ МОДЕЛИ
# =============================================================================

print("\n" + "="*70)
print("3. ОБУЧЕНИЕ МОДЕЛИ")
print("="*70)

# Колбэки
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=3,
        restore_best_weights=True,
        verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=2,
        verbose=1
    )
]

print("\nНачинаю обучение...\n")

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=1
)

print("\n" + "="*70)
print("ОБУЧЕНИЕ ЗАВЕРШЕНО")
print("="*70)

# Финальные метрики
final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
print(f"\nФинальная точность:")
print(f"  Обучающая выборка: {final_train_acc:.4f}")
print(f"  Валидационная выборка: {final_val_acc:.4f}")

# =============================================================================
# 4. ВИЗУАЛИЗАЦИЯ ОБУЧЕНИЯ
# =============================================================================

print("\n" + "="*70)
print("4. ВИЗУАЛИЗАЦИЯ ОБУЧЕНИЯ")
print("="*70)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# График точности
axes[0].plot(history.history['accuracy'], label='Обучающая', linewidth=2)
axes[0].plot(history.history['val_accuracy'], label='Валидационная', linewidth=2)
axes[0].set_title('Точность модели', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Эпоха', fontsize=12)
axes[0].set_ylabel('Точность', fontsize=12)
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# График потерь
axes[1].plot(history.history['loss'], label='Обучающая', linewidth=2)
axes[1].plot(history.history['val_loss'], label='Валидационная', linewidth=2)
axes[1].set_title('Функция потерь', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Эпоха', fontsize=12)
axes[1].set_ylabel('Потери', fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/kaggle/working/training_history.png', dpi=300, bbox_inches='tight')
plt.show()
print("\nГрафик обучения сохранен: /kaggle/working/training_history.png")

# =============================================================================
# 5. ОЦЕНКА НА ТЕСТОВОЙ ВЫБОРКЕ
# =============================================================================

print("\n" + "="*70)
print("5. ОЦЕНКА НА ТЕСТОВОЙ ВЫБОРКЕ")
print("="*70)

# Оценка
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nРезультаты на тестовой выборке:")
print(f"  Loss: {test_loss:.4f}")
print(f"  Accuracy: {test_accuracy:.4f}")
print(f"  Процент правильных предсказаний: {test_accuracy*100:.2f}%")

# Предсказания
print("\nВыполняю предсказания...")
y_pred_probs = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_probs, axis=1)

# Метрики по классам
print("\n" + "-"*70)
print("Детальный отчет по классам:")
print("-"*70)
print(classification_report(y_test, y_pred, target_names=class_names, digits=4))

# =============================================================================
# 6. ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ
# =============================================================================

print("\n" + "="*70)
print("6. ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ")
print("="*70)

# Матрица ошибок
cm = confusion_matrix(y_test, y_pred)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(10, 8))
sns.heatmap(
    cm_normalized,
    annot=True,
    fmt='.2f',
    cmap='Blues',
    xticklabels=class_names,
    yticklabels=class_names,
    cbar_kws={'label': 'Доля'},
    square=True
)
plt.title('Матрица ошибок (нормализованная)', fontsize=14, fontweight='bold')
plt.ylabel('Истинный класс', fontsize=12)
plt.xlabel('Предсказанный класс', fontsize=12)
plt.tight_layout()
plt.savefig('/kaggle/working/confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
print("\nМатрица ошибок сохранена: /kaggle/working/confusion_matrix.png")

# Распределение классов
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

unique, counts = np.unique(y_test, return_counts=True)
axes[0].bar([class_names[i] for i in unique], counts, color='skyblue', edgecolor='black')
axes[0].set_title('Истинное распределение классов', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Класс', fontsize=12)
axes[0].set_ylabel('Количество', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)

unique, counts = np.unique(y_pred, return_counts=True)
axes[1].bar([class_names[i] for i in unique], counts, color='lightcoral', edgecolor='black')
axes[1].set_title('Предсказанное распределение классов', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Класс', fontsize=12)
axes[1].set_ylabel('Количество', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('/kaggle/working/class_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
print("\nРаспределение классов сохранено: /kaggle/working/class_distribution.png")

# =============================================================================
# 7. СОХРАНЕНИЕ МОДЕЛИ
# =============================================================================

print("\n" + "="*70)
print("7. СОХРАНЕНИЕ МОДЕЛИ")
print("="*70)

model.save('/kaggle/working/text_classifier.keras')
print("\nМодель сохранена: /kaggle/working/text_classifier.keras")

# Сохранение информации о классах
np.save('/kaggle/working/class_names.npy', class_names)
print("Названия классов сохранены: /kaggle/working/class_names.npy")

# Сохранение метрик
metrics_data = {
    'test_accuracy': test_accuracy,
    'test_loss': test_loss,
    'final_train_acc': final_train_acc,
    'final_val_acc': final_val_acc,
    'num_texts': len(texts),
    'num_classes': len(class_names),
    'epochs_trained': len(history.history['loss'])
}
np.save('/kaggle/working/metrics.npy', metrics_data)
print("Метрики сохранены: /kaggle/working/metrics.npy")

print("\n" + "="*70)
print("ОБУЧЕНИЕ И ОЦЕНКА ЗАВЕРШЕНЫ")
print("="*70)

print(f"\nОбработано текстов: {len(texts)}")
print(f"Количество классов: {len(class_names)}")
print(f"Классы: {', '.join(class_names)}")
print(f"Обучено эпох: {len(history.history['loss'])}")
print(f"Точность на обучающей выборке: {final_train_acc:.4f}")
print(f"Точность на валидационной выборке: {final_val_acc:.4f}")
print(f"Точность на тестовой выборке: {test_accuracy:.4f}")
```
